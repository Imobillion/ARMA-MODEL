```python
import pandas as pd
import matplotlib.pyplot as plt

# Example time series data
# Read the CSV file into a pandas DataFrame
df = pd.read_csv('holidays_events.csv')
import matplotlib.dates as mdates

plt.figure(figsize=(15, 6))
plt.plot(df['date'], df['locale_name'], marker='o', linestyle='-')

plt.xlabel('Date')
plt.ylabel('Locale Name')
plt.title('Time Series Plot')

# rotate labels
plt.xticks(rotation=90, ha='center')

# show one tick every 7 days (example)
plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))

plt.grid(True)
plt.tight_layout()
plt.show()
```

The time series plot you generated shows the occurrences of holidays and events (locale_name) over time (date).

Here are some findings and insights from the plot:

Concentration of events: There appear to be periods where multiple events occur close together, and other periods with fewer events.

Locale-specific patterns: Different locales (cities or regions) have varying numbers and timings of events. Some locales seem to have more frequent events than others.

Possible seasonality: While not immediately obvious without more detailed analysis, there might be some seasonal patterns in the timing of certain types of events (e.g., holidays).

Data granularity: The plot shows individual events. To see broader trends, you might consider aggregating the data (e.g., counting the number of events per month or year).

To gain deeper insights, you could consider:

Analyzing the type of event to see if certain types follow different patterns.
Investigating the description column for more details about specific events.
Aggregating the data by time period (month, year) to see overall trends in event frequency.
Creating separate plots for different locales or event types.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'date' is datetime
df['date'] = pd.to_datetime(df['date'])

# ------------------------------
# 1. Event type patterns over time
# ------------------------------
plt.figure(figsize=(14, 6))
sns.countplot(data=df, x='type', order=df['type'].value_counts().index)
plt.xticks(rotation=45, ha='right')
plt.title("Event Counts by Type")
plt.tight_layout()
plt.show()

# ------------------------------
# 2. Explore description column
# (Top 20 frequent words in description)
# ------------------------------
from collections import Counter
import re

all_words = " ".join(df['description'].astype(str)).lower()
words = re.findall(r'\w+', all_words)
common_words = Counter(words).most_common(20)

common_df = pd.DataFrame(common_words, columns=['word', 'count'])

plt.figure(figsize=(12, 6))
sns.barplot(data=common_df, x='word', y='count', palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title("Top 20 Words in Event Descriptions")
plt.tight_layout()
plt.show()

# ------------------------------
# 3. Aggregate by month and year
# ------------------------------
df['year_month'] = df['date'].dt.to_period('M')
monthly_counts = df.groupby('year_month').size()

plt.figure(figsize=(14, 6))
monthly_counts.plot(marker='o')
plt.title("Monthly Event Frequency")
plt.xlabel("Month")
plt.ylabel("Number of Events")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# ------------------------------
# 4. Separate plots by locale and event type
# ------------------------------
plt.figure(figsize=(14, 6))
sns.countplot(data=df, x='locale_name', hue='type',
              order=df['locale_name'].value_counts().index[:10])  # top 10 locales
plt.xticks(rotation=45, ha='right')
plt.title("Event Types Across Locales (Top 10)")
plt.tight_layout()
plt.show()

# give explanation with findings and insights

```

```python
# decompose the time series
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# ensure date is datetime
df['date'] = pd.to_datetime(df['date'])

# aggregate events by day (or month if data is sparse)
daily_counts = df.groupby('date').size()

# decompose the time series
decomposition = seasonal_decompose(daily_counts, model='additive', period=30)  # period=30 for monthly seasonality

# plot decomposition
plt.rcParams.update({'figure.figsize': (14, 10)})
decomposition.plot()
plt.suptitle("Time Series Decomposition", fontsize=16)
plt.show()
```

1. Observed

This is the original time series (your monthly or daily event counts).

It shows the raw data with all fluctuations.

Example: You may see spikes in certain months (e.g., more events during the rainy season).

Insight: This is the “real-world” signal before breaking it down.

 2. Trend

A smoothed curve that captures the long-term direction.

Ignores short-term ups and downs, focusing on whether events are increasing, decreasing, or stable over years.

Insights you might see:

If steadily rising → events are happening more often (possible systemic issue or improved detection).

If declining → interventions are working.

If flat → no major long-term changes.

 3. Seasonality

A repeating wave-like pattern that captures cycles.

For monthly data, this might show peaks in specific months (e.g., events higher in July, lower in December).

For weekly data, it could show day-of-week patterns.

Insights you might see:

Spikes in certain months = seasonal dependency (e.g., weather-related, festival-related).

Stable repeating cycles confirm that events are predictable by time of year.

4. Residual (Noise)

Whatever is left after removing trend + seasonality.

Shows irregular events, anomalies, or random fluctuations.

These are often sudden spikes/drops caused by rare events (e.g., natural disasters, one-off system failures).

Insights you might see:

If residuals are small → trend + seasonality explain most of the variation (systematic).

If residuals are large → many unexpected events (less predictable).

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(daily_counts.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])

if result[1] < 0.05:
    print("✅ The series is stationary")
else:
    print("⚠️ The series is not stationary, differencing may be required")
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Example: ARMA(2,1) → change (p,q) after checking ACF/PACF
model = ARIMA(daily_counts, order=(2,0,1))  # (p,d,q) where d=0 for ARMA
arma_result = model.fit()

print(arma_result.summary())
```

```python
arma_result.plot_diagnostics(figsize=(12,6))
plt.show()
```

```python
forecast = arma_result.get_forecast(steps=30)  # next 30 days
forecast_ci = forecast.conf_int()

plt.figure(figsize=(12,6))
plt.plot(daily_counts, label='Observed')
plt.plot(forecast.predicted_mean, label='Forecast', color='red')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:, 0],
                 forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.legend()
plt.title("ARMA Forecast")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import joblib  # for saving model

# Ensure datetime index
df['date'] = pd.to_datetime(df['date'])
daily_counts = df.groupby('date').size()
daily_counts = daily_counts.asfreq('D').fillna(0)  # ensure daily frequency

# Split into train and test
train_size = int(len(daily_counts) * 0.8)
train, test = daily_counts[:train_size], daily_counts[train_size:]

plt.figure(figsize=(12,6))
plt.plot(train, label="Train")
plt.plot(test, label="Test")
plt.title("Train-Test Split")
plt.legend()
plt.show()
```

```python
# Example ARMA(2,1) → adjust p and q
p, q = 2, 1
arma_model = ARIMA(train, order=(p,0,q))
arma_result = arma_model.fit()

print(arma_result.summary())
```

```python
forecast = arma_result.get_forecast(steps=len(test))
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

plt.figure(figsize=(12,6))
plt.plot(train, label='Train')
plt.plot(test, label='Test', color='orange')
plt.plot(forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:, 0],
                 forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.title("ARMA Train-Test Forecast")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

mse = mean_squared_error(test, forecast_mean)
mae = mean_absolute_error(test, forecast_mean)
rmse = np.sqrt(mse)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# Save model to file
joblib.dump(arma_result, "arma_model.pkl")

# Load it back later
loaded_model = joblib.load("arma_model.pkl")
print("✅ Model saved and loaded successfully!")
```

```python
import pandas as pd
import numpy as np
import warnings
from statsmodels.tsa.arima.model import ARIMA

# Ignore warnings for clean output
warnings.filterwarnings("ignore")

# Ensure datetime index
df['date'] = pd.to_datetime(df['date'])
daily_counts = df.groupby('date').size()
daily_counts = daily_counts.asfreq('D').fillna(0)

# Train-test split (80/20)
train_size = int(len(daily_counts) * 0.8)
train, test = daily_counts[:train_size], daily_counts[train_size:]

# Function to find best ARMA order
def select_arma_order(train, p_range=5, q_range=5, criterion="aic"):
    best_score, best_order = float("inf"), None
    results = []

    for p in range(p_range + 1):
        for q in range(q_range + 1):
            if p == 0 and q == 0:  # skip ARMA(0,0)
                continue
            try:
                model = ARIMA(train, order=(p,0,q))
                result = model.fit()
                score = getattr(result, criterion)
                results.append(((p, q), score))
                if score < best_score:
                    best_score, best_order = score, (p, q)
            except Exception as e:
                continue

    results = sorted(results, key=lambda x: x[1])
    return best_order, results

# Run order selection
best_order, all_results = select_arma_order(train, p_range=5, q_range=5, criterion="aic")
print("✅ Best ARMA order based on AIC:", best_order)

# Top 5 candidates
print("\nTop 5 ARMA orders by AIC:")
for order, score in all_results[:5]:
    print(f"ARMA{order} → AIC: {score:.2f}")

# Fit final best model
best_model = ARIMA(train, order=(best_order[0],0,best_order[1])).fit()

# Forecast on test set
forecast = best_model.get_forecast(steps=len(test))
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Plot forecast vs. actual
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(train, label="Train")
plt.plot(test, label="Test", color='orange')
plt.plot(forecast_mean, label=f"ARMA{best_order} Forecast", color='red')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:,0], forecast_ci.iloc[:,1],
                 color='pink', alpha=0.3)
plt.title(f"ARMA{best_order} Forecast (Auto-selected by AIC)")
plt.legend()
plt.show()
```
